```python
 import numpy as np
 import pandas as pd
 from pylab import mpl, plt
 plt.style.use('seaborn-v0_8')
 mpl.rcParams['font.family'] = 'serif'
 %matplotlib inline
 from io import StringIO
 import math as math
 import numpy.random as npr
 import scipy.stats as scs
```

```python
values = data['clot']
values.plot()
```

```python
E=values.mean()
```

```python
#MouvementGeoBrow
S0 
r = E
sigma =values.std()
T = 0.01
I = 100
ST1 = S0 * np.exp((r - 0.5 * sigma ** 2) * T +
sigma * math.sqrt(T) * npr.standard_normal(I))
```

```python
ST1
```

```python
plt.figure(figsize=(10, 6))
plt.hist(ST1, bins=50)
plt.xlabel('Repartition possibilits du prix du ptrole')
plt.ylabel('frequency');
```

```python
#Echelle logarithmique
ST2 = S0 * npr.lognormal((r - 0.5 * sigma ** 2) * T, sigma * math.sqrt(T), size=I)
plt.figure(figsize=(10, 6))
plt.hist(ST2, bins=50)
plt.xlabel('index level')
plt.ylabel('frequency');
```

```python
#Fonction qui permet extraire statistiques importantes

def print_statistics(a1, a2):
 ''' Prints selected statistics.
 Parameters
 ==========
 a1, a2: ndarray objects
 results objects from simulation
 '''
 sta1 = scs.describe(a1) 
 sta2 = scs.describe(a2) 
 print('%14s %14s %14s' %
 ('statistic', 'data set 1', 'data set 2'))
 print(45 * "-")
 print('%14s %14.3f %14.3f' % ('size', sta1[0], sta2[0]))
 print('%14s %14.3f %14.3f' % ('min', sta1[1][0], sta2[1][0]))
 print('%14s %14.3f %14.3f' % ('max', sta1[1][1], sta2[1][1]))
 print('%14s %14.3f %14.3f' % ('mean', sta1[2], sta2[2]))
 print('%14s %14.3f %14.3f' % ('std', np.sqrt(sta1[3]), np.sqrt(sta2[3])))
 print('%14s %14.3f %14.3f' % ('skew', sta1[4], sta2[4]))
 print('%14s %14.3f %14.3f' % ('kurtosis', sta1[5], sta2[5]))
```

```python
print_statistics(ST1, ST2)
```

```python
#Mouvement brownien geometrique ou ce n'est plus So qui est considr mais St-1
S0 
r = E
sigma =values.std()
I = 10000 
M = 50 
dt = T / M 
S = np.zeros((M + 1, I)) 
S[0] = S0 
for t in range(1, M + 1):
    S[t] = S[t - 1] * np.exp((r - 0.5 * sigma ** 2) * dt +sigma * math.sqrt(dt) * npr.standard_normal(I))
```

```python
plt.figure(figsize=(10, 6))
plt.hist(S[1], bins=50)
plt.xlabel('prix petrole euro')
plt.ylabel('frequency')
```
